```python
import itertools
import matplotlib.pyplot as plt

# Puzzle definition
words = ['EAT', 'THAT']
result = 'APPLE'
all_letters = sorted(set(''.join(words) + result))
assert len(all_letters) <= 10, "Too many unique letters!"

# Converts word to its numeric value using the mapping
def to_number(word, letter_map):
    return int(''.join(str(letter_map[c]) for c in word))

# Solving with permutations
def solve_cryptarithmetic():
    digits = range(10)
    for perm in itertools.permutations(digits, len(all_letters)):
        letter_map = dict(zip(all_letters, perm))

        # Skip if any leading letter is zero
        if any(letter_map[word[0]] == 0 for word in words + [result]):
            continue

        eat = to_number('EAT', letter_map)
        that = to_number('THAT', letter_map)
        apple = to_number('APPLE', letter_map)

        if eat + that == apple:
            return letter_map, eat, that, apple
    return None

# Visualization function
def visualize_solution(letter_map, eat, that, apple):
    fig, ax = plt.subplots(figsize=(8, 4))
    ax.axis('off')

    def show_text(y, text):
        ax.text(0.1, y, text, fontsize=20, family='monospace')

    show_text(0.8, f"  {eat:>5}")
    show_text(0.6, f"+ {that:>5}")
    show_text(0.5, "-" * 8)
    show_text(0.3, f" {apple:>6}")
    show_text(0.1, f"Letter → Digit: " + ', '.join(f"{k}={v}" for k, v in sorted(letter_map.items())))

    plt.title("CRYPTARITHMETIC: EAT + THAT = APPLE", fontsize=14)
    plt.show()

# Run and display solution
solution = solve_cryptarithmetic()
if solution:
    letter_map, eat, that, apple = solution
    print(f"✔️ Solution found: {eat} + {that} = {apple}")
    print("Mapping:", letter_map)
    visualize_solution(letter_map, eat, that, apple)
else:
    print("❌ No solution found.")
```
