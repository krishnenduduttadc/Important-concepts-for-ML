```python
from graphviz import Digraph

dot = Digraph()

dot.node('A', 'Start')
dot.node('B', 'Process')
dot.node('C', 'Decision')
dot.node('D', 'End')

dot.edges(['AB', 'BC'])
dot.edge('C', 'D', label='Yes')
dot.edge('C', 'A', label='No')

dot.render('simple_flowchart', format='png', view=True)  # Saves and opens image
dot  # Displays in notebook

```

```python
org = Digraph(comment='Org Chart')

org.node('CEO')
org.node('CTO')
org.node('CFO')
org.node('Engineer')
org.node('Accountant')

org.edges([('CEO', 'CTO'), ('CEO', 'CFO'), ('CTO', 'Engineer'), ('CFO', 'Accountant')])

org
```

```python
tree = Digraph()

tree.node('1', 'Root')
tree.node('2', 'Left')
tree.node('3', 'Right')

tree.edge('1', '2')
tree.edge('1', '3')

tree
```

```python
fsm = Digraph()

fsm.attr(rankdir='LR')  # Left to right orientation

fsm.node('S0', shape='circle')
fsm.node('S1', shape='circle')
fsm.node('S2', shape='doublecircle')

fsm.edge('S0', 'S1', label='a')
fsm.edge('S1', 'S2', label='b')
fsm.edge('S2', 'S0', label='c')

fsm
```
