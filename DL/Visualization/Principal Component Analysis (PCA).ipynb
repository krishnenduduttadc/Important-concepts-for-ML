```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Generate correlated 2D data
np.random.seed(0)
X = np.random.multivariate_normal([0, 0], [[3, 2.5], [2.5, 2]], 200)

# Apply PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
components = pca.components_          # principal axes
explained_var = pca.explained_variance_

# Project data onto first principal component
X_proj = np.dot(X - pca.mean_, components.T)

# Plot original data
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], alpha=0.3, label="Original Data")

# Plot principal axes
origin = pca.mean_
for length, vector in zip(explained_var, components):
    v = vector * 3 * np.sqrt(length)
    plt.quiver(*origin, *v, angles='xy', scale_units='xy', scale=1, color='r', width=0.01)

# Plot projection of data onto first PC
proj_on_pc1 = np.outer(X_pca[:, 0], components[0]) + origin
plt.scatter(proj_on_pc1[:, 0], proj_on_pc1[:, 1], alpha=0.4, color='green', label='Projection on PC1')

plt.axis('equal')
plt.grid(True)
plt.title("PCA: Principal Components & Projection")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.tight_layout()
plt.show()
```
