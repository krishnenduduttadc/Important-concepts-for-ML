```python
#Given an input of size 32 × 32 × 3 and a convolutional layer with 16 filters of size 3 × 3, padding 1, and stride 1,what is the output size of the layer?
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle, Circle
from matplotlib.lines import Line2D
import numpy as np

def draw_conv_network():
    fig, ax = plt.subplots(figsize=(12, 6))
    ax.set_xlim(0, 10)
    ax.set_ylim(0, 6)
    ax.axis('off')
    
    # Title
    plt.title('Convolutional Neural Network Diagram\nInput: 32×32×3 → Conv: 16 filters of 3×3 (padding=1, stride=1)', 
              fontsize=12, pad=20)
    
    # Draw input volume (32×32×3)
    draw_volume(ax, 1, 3, 2, 2, 'Input\n32×32×3', color='skyblue', depth=3)
    
    # Draw convolutional layer
    draw_volume(ax, 5, 3, 2, 2, 'Conv Layer\n16×32×32', color='lightgreen', depth=16, max_display=8)
    
    # Draw arrows between layers
    ax.arrow(3.5, 3, 1, 0, head_width=0.2, head_length=0.1, fc='k', ec='k')
    ax.text(4.25, 3.5, '16 filters\n3×3, pad=1\nstride=1', fontsize=9, ha='center')
    
    plt.tight_layout()
    plt.show()

def draw_volume(ax, x, y, width, height, label, color, depth, max_display=4):
    """Draw a 3D-ish representation of a volume"""
    cube_depth = min(depth, max_display)
    
    # Draw front face
    ax.add_patch(Rectangle((x, y), width, height, fill=True, color=color, ec='k', alpha=0.7))
    ax.text(x + width/2, y + height + 0.2, label, ha='center', va='bottom')
    
    # Draw side faces to give 3D effect
    for i in range(1, cube_depth + 1):
        offset = i * 0.1
        ax.add_patch(Rectangle((x + offset, y - offset), width, height, 
                     fill=False, ec='k', alpha=0.3))
    
    # If there are more channels than we can display, indicate with ellipsis
    if depth > max_display:
        ax.text(x + width + 0.5, y + height/2, f'+{depth-max_display} more...', 
                fontsize=8, va='center')

# Display the diagram
draw_conv_network()


```

```python
#What is the output size of a max-pooling layer with a filter size of 2 × 2 and stride 2 when applied to an input ofsize 28 × 28 × 64
```

```python

import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Create a figure and axis
fig, ax = plt.subplots(figsize=(10, 8))

# Define the layer sizes
input_neurons = 10
hidden_neurons = 6
output_neurons = 3

# Define the positions for the layers
input_layer = [(1, i) for i in range(input_neurons)]
hidden_layer = [(2, i) for i in range(hidden_neurons)]
output_layer = [(3, i) for i in range(output_neurons)]

# Plot input layer neurons
for (x, y) in input_layer:
    ax.plot(x, y, 'bo', markersize=10)
    ax.text(x + 0.1, y, f'Input {y+1}', fontsize=10)

# Plot hidden layer neurons
for (x, y) in hidden_layer:
    ax.plot(x, y, 'go', markersize=10)
    ax.text(x + 0.1, y, f'Hidden {y+1}', fontsize=10)

# Plot output layer neurons
for (x, y) in output_layer:
    ax.plot(x, y, 'ro', markersize=10)
    ax.text(x + 0.1, y, f'Output {y+1}', fontsize=10)

# Draw connections between layers
for (x1, y1) in input_layer:
    for (x2, y2) in hidden_layer:
        ax.plot([x1, x2], [y1, y2], 'k-', lw=0.5)

for (x2, y2) in hidden_layer:
    for (x3, y3) in output_layer:
        ax.plot([x2, x3], [y2, y3], 'k-', lw=0.5)

# Add labels and titles
ax.set_xlim(0, 4)
ax.set_ylim(-1, max(input_neurons, hidden_neurons, output_neurons) + 1)
ax.axis('off')
ax.set_title('Simple Neural Network Diagram')

# Show the plot
plt.show()

```

```python

#Consider a CNN with an input of size 128x128 and a filter of size 3x3 with stride 1 and padding 1. What is theoutput dimension after applying this filter
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def draw_cnn():
    fig, ax = plt.subplots(figsize=(10, 6))

    # Input Image (128x128)
    input_box = patches.Rectangle((0, 1), 2, 2, linewidth=1.5, edgecolor='blue', facecolor='lightblue', label='Input\n128x128')
    ax.add_patch(input_box)
    ax.text(0.5, 2.2, 'Input\n128x128', fontsize=10, color='blue')

    # Conv Filter
    filter_box = patches.Rectangle((2.5, 1.5), 0.5, 0.5, linewidth=1, edgecolor='black', facecolor='orange')
    ax.add_patch(filter_box)
    ax.text(2.2, 2.2, 'Conv 3x3\nStride 1\nPadding 1', fontsize=9, color='darkorange')

    # Feature Map (Output)
    output_box = patches.Rectangle((4, 1), 2, 2, linewidth=1.5, edgecolor='green', facecolor='lightgreen')
    ax.add_patch(output_box)
    ax.text(4.5, 2.2, 'Output\n128x128', fontsize=10, color='green')

    # Arrows to indicate flow
    ax.annotate('', xy=(2, 2), xytext=(2.5, 2.25),
                arrowprops=dict(arrowstyle='->', lw=1.5, color='gray'))
    ax.annotate('', xy=(3, 2), xytext=(4, 2),
                arrowprops=dict(arrowstyle='->', lw=1.5, color='gray'))

    # Clean up plot
    ax.set_xlim(-1, 7)
    ax.set_ylim(0.5, 3.5)
    ax.axis('off')
    ax.set_title('CNN Layer: 128x128 Input → Conv(3x3, stride=1, padding=1) → 128x128 Output', fontsize=12)
    plt.show()

draw_cnn()
```

```python
# Input:
# 100x100x1
# (grayscale image)
# Conv layer:
# 10
# filters of
# 3x3, stride1, padding1.
# Max Pooling:
# 2x2, stride2.
# What is the output shape after these two layer
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def draw_simple_cnn():
    fig, ax = plt.subplots(figsize=(12, 6))

    # Input layer (100x100x1)
    input_rect = patches.Rectangle((0, 2), 1, 1, linewidth=1.5, edgecolor='blue', facecolor='lightblue')
    ax.add_patch(input_rect)
    ax.text(0.1, 3.1, 'Input\n100x100x1', fontsize=10, color='blue')

    # Conv layer (100x100x10)
    for i in range(10):
        conv_rect = patches.Rectangle((2 + i*0.1, 1.8 - i*0.02), 1, 1, linewidth=1, edgecolor='orange', facecolor='moccasin')
        ax.add_patch(conv_rect)
    ax.text(2.2, 3.1, 'Conv Layer\n100x100x10', fontsize=10, color='darkorange')

    # Max Pooling layer (50x50x10)
    for i in range(10):
        pool_rect = patches.Rectangle((4 + i*0.1, 1.6 - i*0.02), 0.8, 0.8, linewidth=1, edgecolor='green', facecolor='lightgreen')
        ax.add_patch(pool_rect)
    ax.text(4.2, 2.9, 'Max Pooling\n50x50x10', fontsize=10, color='green')

    # Arrows between layers
    ax.annotate('', xy=(1, 2.5), xytext=(2, 2.5),
                arrowprops=dict(arrowstyle='->', lw=1.5))
    ax.annotate('', xy=(3, 2.2), xytext=(4, 2.2),
                arrowprops=dict(arrowstyle='->', lw=1.5))

    # Cleanup plot
    ax.set_xlim(-0.5, 7)
    ax.set_ylim(0.5, 3.5)
    ax.axis('off')
    ax.set_title('CNN: 100x100x1 → Conv(3x3x10) → 100x100x10 → MaxPool(2x2) → 50x50x10', fontsize=12)

    plt.show()

draw_simple_cnn()
```

```python
# A CNN layer has:
# Input channels:
# 32
# Output channels:
# 64
# Kernel size:
# (3, 3)
# Bias:
# True
# How many trainable parameters does this layer have?
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def plot_conv_layer():
    fig, ax = plt.subplots(figsize=(12, 6))

    # Input feature maps (32 channels)
    for i in range(32):
        rect = patches.Rectangle((0 + i*0.05, 1.5 - i*0.01), 1, 1, linewidth=1,
                                 edgecolor='blue', facecolor='lightblue', alpha=0.6)
        ax.add_patch(rect)
    ax.text(0.3, 2.6, 'Input Feature Maps\n32 channels', fontsize=10, color='blue')

    # Conv Layer block
    conv_box = patches.Rectangle((2, 1.3), 1.2, 1.2, linewidth=2,
                                 edgecolor='black', facecolor='orange')
    ax.add_patch(conv_box)
    ax.text(2.05, 2.65, 'Conv Layer\n64 filters\n3x3 kernel\nBias=True',
            fontsize=9, color='darkred')

    # Output feature maps (64 channels)
    for i in range(64):
        rect = patches.Rectangle((4 + i*0.03, 1.5 - i*0.005), 0.8, 0.8, linewidth=1,
                                 edgecolor='green', facecolor='lightgreen', alpha=0.6)
        ax.add_patch(rect)
    ax.text(4.2, 2.6, 'Output Feature Maps\n64 channels', fontsize=10, color='green')

    # Arrows
    ax.annotate('', xy=(1.6, 2.0), xytext=(2, 2.0),
                arrowprops=dict(arrowstyle='->', lw=1.5))
    ax.annotate('', xy=(3.2, 2.0), xytext=(4, 2.0),
                arrowprops=dict(arrowstyle='->', lw=1.5))

    # Final setup
    ax.set_xlim(-0.5, 7)
    ax.set_ylim(0.8, 3)
    ax.axis('off')
    ax.set_title('CNN Layer: 32 Input Channels → Conv(3x3, 64 filters, bias=True) → 64 Output Channels', fontsize=12)

    plt.show()

plot_conv_layer()
```

```python
# A CNN layer processes an input of size
# (64, 64, 16) with 32 filters of size (7, 7), stride 1 and no padding. How manymultiplications are required for this layer?
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def plot_cnn_7x7():
    fig, ax = plt.subplots(figsize=(14, 6))

    # Input feature maps (16 channels)
    for i in range(16):
        rect = patches.Rectangle((0 + i * 0.08, 1.5 - i * 0.01), 1, 1, linewidth=1,
                                 edgecolor='blue', facecolor='lightblue', alpha=0.6)
        ax.add_patch(rect)
    ax.text(0.2, 2.7, 'Input\n64x64x16', fontsize=10, color='blue')

    # Conv Layer (visual marker)
    conv_layer = patches.Rectangle((2.5, 1.3), 1.2, 1.2, linewidth=2,
                                   edgecolor='black', facecolor='orange')
    ax.add_patch(conv_layer)
    ax.text(2.55, 2.65, 'Conv Layer\n32 filters\n7x7 kernel\nStride=1\nPadding=0',
            fontsize=9, color='darkred')

    # Output feature maps (32 channels)
    for i in range(32):
        rect = patches.Rectangle((4.5 + i * 0.04, 1.4 - i * 0.005), 0.8, 0.8, linewidth=1,
                                 edgecolor='green', facecolor='lightgreen', alpha=0.6)
        ax.add_patch(rect)
    ax.text(4.8, 2.5, 'Output\n58x58x32', fontsize=10, color='green')

    # Arrows
    ax.annotate('', xy=(1.9, 2.0), xytext=(2.5, 2.0),
                arrowprops=dict(arrowstyle='->', lw=1.5))
    ax.annotate('', xy=(3.7, 2.0), xytext=(4.5, 2.0),
                arrowprops=dict(arrowstyle='->', lw=1.5))

    # Setup
    ax.set_xlim(-0.5, 8)
    ax.set_ylim(0.8, 3)
    ax.axis('off')
    ax.set_title('CNN Layer: 64x64x16 Input → Conv(7x7, 32 filters, stride=1, padding=0) → 58x58x32 Output', fontsize=12)

    plt.show()

plot_cnn_7x7()
```

```python
# Input:
# (32, 32, 3)

# Architecture:
# Two Conv2D layers:
# 1. 32 filters of (5, 5) stride 1 , padding 2

# 2. 64 filters of (3, 3) , stride 2, padding 0.
# What is the total number of trainable parameters
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def plot_two_conv_layers():
    fig, ax = plt.subplots(figsize=(16, 6))

    # Input (32x32x3)
    for i in range(3):
        rect = patches.Rectangle((0 + i * 0.15, 1.8 - i * 0.02), 1, 1, linewidth=1.2,
                                 edgecolor='blue', facecolor='lightblue', alpha=0.7)
        ax.add_patch(rect)
    ax.text(0.2, 3.1, 'Input\n32x32x3', fontsize=10, color='blue')

    # Conv Layer 1 (32x32x32)
    for i in range(6):  # showing 6 of 32 channels
        rect = patches.Rectangle((2 + i * 0.1, 1.6 - i * 0.01), 0.9, 0.9, linewidth=1,
                                 edgecolor='orange', facecolor='moccasin', alpha=0.7)
        ax.add_patch(rect)
    ax.text(2.2, 2.9, 'Conv2D\n32 filters\n5x5, stride=1, pad=2\n→ 32x32x32', fontsize=9, color='darkorange')

    # Conv Layer 2 (15x15x64)
    for i in range(6):  # showing 6 of 64 channels
        rect = patches.Rectangle((4.5 + i * 0.1, 1.4 - i * 0.005), 0.7, 0.7, linewidth=1,
                                 edgecolor='green', facecolor='lightgreen', alpha=0.7)
        ax.add_patch(rect)
    ax.text(4.7, 2.6, 'Conv2D\n64 filters\n3x3, stride=2, pad=0\n→ 15x15x64', fontsize=9, color='darkgreen')

    # Arrows
    ax.annotate('', xy=(1.4, 2.2), xytext=(2, 2.2),
                arrowprops=dict(arrowstyle='->', lw=1.5))
    ax.annotate('', xy=(3.5, 2.1), xytext=(4.5, 2.1),
                arrowprops=dict(arrowstyle='->', lw=1.5))

    # Final setup
    ax.set_xlim(-0.5, 7)
    ax.set_ylim(1, 3.3)
    ax.axis('off')
    ax.set_title('CNN: 32x32x3 Input → Conv(5x5x32, stride=1, pad=2) → 32x32x32 → Conv(3x3x64, stride=2, pad=0) → 15x15x64', fontsize=12)

    plt.show()

plot_two_conv_layers()
```
