```python
# Create a map centered around India
m = folium.Map(location=[20.5937, 78.9629], zoom_start=5)

# Coordinates for the cities
kolkata = [22.5726, 88.3639]
patna = [25.5941, 85.1376]
delhi = [28.6139, 77.2090]
mumbai = [19.0760, 72.8777]

# Add markers for the cities
folium.Marker(kolkata, popup="Kolkata").add_to(m)
folium.Marker(patna, popup="Patna").add_to(m)
folium.Marker(delhi, popup="Delhi").add_to(m)
folium.Marker(mumbai, popup="Mumbai").add_to(m)

# Draw lines between the cities
folium.PolyLine([kolkata, patna], color="blue", weight=2.5, opacity=1).add_to(m)
folium.PolyLine([patna, delhi], color="blue", weight=2.5, opacity=1).add_to(m)
folium.PolyLine([delhi, mumbai], color="blue", weight=2.5, opacity=1).add_to(m)




lines = [
    {"type": "Feature", "geometry": {"type": "LineString", "coordinates": [kolkata, patna]}},
    {"type": "Feature", "geometry": {"type": "LineString", "coordinates": [patna, delhi]}},
    {"type": "Feature", "geometry": {"type": "LineString", "coordinates": [delhi, mumbai]}}
]

# Create a GeoJSON object
geojson_data = {
    "type": "FeatureCollection",
    "features": lines
}

# Write the GeoJSON data to a file
with open('india_cities_lines.geojson', 'w') as f:
    json.dump(geojson_data, f)

# Display the map
m
```
