```python
#Probability & Sampling
```

```python
import tensorflow as tf
import matplotlib.pyplot as plt

# Set a seed for reproducibility
tf.random.set_seed(42)

# Sample sizes
n = 1000

# 1. Normal Distribution
normal_samples = tf.random.normal([n], mean=0.0, stddev=1.0)
plt.figure(figsize=(10, 6))
plt.hist(normal_samples.numpy(), bins=50, density=True, alpha=0.7, color='skyblue')
plt.title("Normal Distribution (mean=0, stddev=1)")
plt.xlabel("Value")
plt.ylabel("Density")
plt.grid(True)
plt.show()

# 2. Uniform Distribution
uniform_samples = tf.random.uniform([n], minval=0, maxval=1)
plt.figure(figsize=(10, 6))
plt.hist(uniform_samples.numpy(), bins=50, density=True, alpha=0.7, color='green')
plt.title("Uniform Distribution (0 to 1)")
plt.xlabel("Value")
plt.ylabel("Density")
plt.grid(True)
plt.show()

# 3. Poisson Distribution (discrete)
poisson_samples = tf.random.poisson(shape=[n], lam=5.0)
plt.figure(figsize=(10, 6))
plt.hist(poisson_samples.numpy(), bins=range(0, 20), density=True, alpha=0.7, color='orange', align='left')
plt.title("Poisson Distribution ( = 5)")
plt.xlabel("Value")
plt.ylabel("Probability")
plt.grid(True)
plt.show()

# 4. Gamma Distribution
gamma_samples = tf.random.gamma(shape=[n], alpha=2.0)
plt.figure(figsize=(10, 6))
plt.hist(gamma_samples.numpy(), bins=50, density=True, alpha=0.7, color='purple')
plt.title("Gamma Distribution ( = 2)")
plt.xlabel("Value")
plt.ylabel("Density")
plt.grid(True)
plt.show()
```

```python
#statistical operations
```

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np

# Generate random data (normal distribution)
tf.random.set_seed(42)
data = tf.random.normal([1000], mean=5.0, stddev=2.0)

# Convert to NumPy for plotting
data_np = data.numpy()

# Calculate basic statistics
mean = tf.reduce_mean(data).numpy()
std = tf.math.reduce_std(data).numpy()
variance = tf.math.reduce_variance(data).numpy()
min_val = tf.reduce_min(data).numpy()
max_val = tf.reduce_max(data).numpy()
sum_val = tf.reduce_sum(data).numpy()
argmax_idx = tf.math.argmax(data).numpy()
argmin_idx = tf.math.argmin(data).numpy()

# Print values
print(f"Mean: {mean:.3f}")
print(f"Standard Deviation: {std:.3f}")
print(f"Variance: {variance:.3f}")
print(f"Min: {min_val:.3f}")
print(f"Max: {max_val:.3f}")
print(f"Sum: {sum_val:.3f}")
print(f"Argmax Index: {argmax_idx}, Value: {data[argmax_idx].numpy():.3f}")
print(f"Argmin Index: {argmin_idx}, Value: {data[argmin_idx].numpy():.3f}")

# Plot histogram with mean and std lines
plt.figure(figsize=(10, 6))
plt.hist(data_np, bins=50, density=True, alpha=0.6, color='skyblue', label='Data Histogram')
plt.axvline(mean, color='red', linestyle='--', label=f'Mean = {mean:.2f}')
plt.axvline(mean + std, color='green', linestyle='--', label=f'+1 Std = {mean + std:.2f}')
plt.axvline(mean - std, color='green', linestyle='--', label=f'-1 Std = {mean - std:.2f}')
plt.axvline(min_val, color='black', linestyle=':', label='Min')
plt.axvline(max_val, color='purple', linestyle=':', label='Max')
plt.title("Statistical Operations on TensorFlow Data")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Optimization algorithms
```

```python
import tensorflow as tf
import matplotlib.pyplot as plt

# Define the loss function: L(w) = (w - 3)^2
def loss_fn(w):
    return (w - 3.0) ** 2

# Learning rate
lr = 0.1
steps = 50

# Initialize two weights for comparison
w_sgd = tf.Variable(0.0)
w_adam = tf.Variable(0.0)

# Track weight updates for plotting
history_sgd = []
history_adam = []

# Optimizers
optimizer_sgd = tf.keras.optimizers.SGD(learning_rate=lr)
optimizer_adam = tf.keras.optimizers.Adam(learning_rate=lr)

# Run optimization
for step in range(steps):
    with tf.GradientTape() as tape:
        loss = loss_fn(w_sgd)
    grad = tape.gradient(loss, [w_sgd])
    optimizer_sgd.apply_gradients(zip(grad, [w_sgd]))
    history_sgd.append(w_sgd.numpy())

    with tf.GradientTape() as tape:
        loss = loss_fn(w_adam)
    grad = tape.gradient(loss, [w_adam])
    optimizer_adam.apply_gradients(zip(grad, [w_adam]))
    history_adam.append(w_adam.numpy())

# Plotting the optimization path
plt.figure(figsize=(10, 6))
plt.plot(history_sgd, label='SGD', marker='o')
plt.plot(history_adam, label='Adam', marker='x')
plt.axhline(y=3.0, color='red', linestyle='--', label='Minimum at w = 3')
plt.title('Optimization Progress of SGD vs Adam')
plt.xlabel('Step')
plt.ylabel('w')
plt.grid(True)
plt.legend()
plt.show()
```

```python
#Activation function
```

```python


import numpy as np
import matplotlib.pyplot as plt

# Define the input range
x = np.linspace(-10, 10, 400)

# Define the activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

def softmax(x):
    e_x = np.exp(x - np.max(x))  # Subtract max for numerical stability
    return e_x / e_x.sum(axis=0)

# Compute the values of the activation functions
sigmoid_values = sigmoid(x)
tanh_values = tanh(x)
relu_values = relu(x)
leaky_relu_values = leaky_relu(x)
softmax_values = softmax(x)  # Softmax works with vector inputs, so we'll treat x as a vector

# Plot the activation functions
plt.figure(figsize=(12, 10))

# Sigmoid Plot
plt.subplot(3, 2, 1)
plt.plot(x, sigmoid_values, label='Sigmoid', color='blue')
plt.title("Sigmoid Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.legend()

# Tanh Plot
plt.subplot(3, 2, 2)
plt.plot(x, tanh_values, label='Tanh', color='green')
plt.title("Tanh Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.legend()

# ReLU Plot
plt.subplot(3, 2, 3)
plt.plot(x, relu_values, label='ReLU', color='red')
plt.title("ReLU Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.legend()

# Leaky ReLU Plot
plt.subplot(3, 2, 4)
plt.plot(x, leaky_relu_values, label='Leaky ReLU', color='purple')
plt.title("Leaky ReLU Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.legend()

# Softmax Plot (for a vector, let's just plot the output of Softmax as a function of the input)
plt.subplot(3, 2, 5)
plt.plot(x, softmax_values, label='Softmax', color='orange')
plt.title("Softmax Activation Function")
plt.xlabel("Input (treated as vector)")
plt.ylabel("Output")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()



```

```python
#Algeblic operations
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define input arrays (vectors/matrices)
x = np.linspace(0, 10, 100)  # Create a vector from 0 to 10 with 100 points
y = np.linspace(10, 0, 100)  # Create another vector (reverse of x)

# Perform algebraic operations
addition = x + y  # Element-wise addition
subtraction = x - y  # Element-wise subtraction
multiplication = x * y  # Element-wise multiplication (Hadamard product)
division = x / (y + 1e-10)  # Element-wise division (avoid division by zero)
exponentiation = np.power(x, 2)  # Element-wise exponentiation (x^2)

# Plotting the algebraic operations
plt.figure(figsize=(12, 10))

# Addition Plot
plt.subplot(3, 2, 1)
plt.plot(x, addition, label='Addition (x + y)', color='blue')
plt.title("Algebraic Addition")
plt.xlabel("x")
plt.ylabel("x + y")
plt.grid(True)
plt.legend()

# Subtraction Plot
plt.subplot(3, 2, 2)
plt.plot(x, subtraction, label='Subtraction (x - y)', color='green')
plt.title("Algebraic Subtraction")
plt.xlabel("x")
plt.ylabel("x - y")
plt.grid(True)
plt.legend()

# Multiplication Plot
plt.subplot(3, 2, 3)
plt.plot(x, multiplication, label='Multiplication (x * y)', color='red')
plt.title("Algebraic Multiplication")
plt.xlabel("x")
plt.ylabel("x * y")
plt.grid(True)
plt.legend()

# Division Plot
plt.subplot(3, 2, 4)
plt.plot(x, division, label='Division (x / y)', color='purple')
plt.title("Algebraic Division")
plt.xlabel("x")
plt.ylabel("x / y")
plt.grid(True)
plt.legend()

# Exponentiation Plot
plt.subplot(3, 2, 5)
plt.plot(x, exponentiation, label='Exponentiation (x^2)', color='orange')
plt.title("Algebraic Exponentiation")
plt.xlabel("x")
plt.ylabel("x^2")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
