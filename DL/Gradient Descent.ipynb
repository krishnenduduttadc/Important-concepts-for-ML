```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function f(x) = x^2 + 4x + 4
def func(x):
    return x**2 + 4*x + 4

# Define the derivative of the function f'(x) = 2x + 4
def grad(x):
    return 2*x + 4

# Gradient descent implementation
def gradient_descent(learning_rate, initial_x, iterations):
    x = initial_x
    history = [x]
    
    for _ in range(iterations):
        # Compute the gradient (derivative)
        gradient = grad(x)
        
        # Update x by moving in the direction of the negative gradient
        x = x - learning_rate * gradient
        history.append(x)
    
    return history

# Parameters
learning_rate = 0.1
initial_x = 5  # Start at x = 5
iterations = 50

# Run gradient descent
history = gradient_descent(learning_rate, initial_x, iterations)

# Visualize the function and the gradient descent steps
x_vals = np.linspace(-6, 2, 400)
y_vals = func(x_vals)

plt.figure(figsize=(10, 6))

# Plot the function f(x)
plt.plot(x_vals, y_vals, label="f(x) = x^2 + 4x + 4", color='b')

# Plot the gradient descent steps
history_vals = [func(x) for x in history]
plt.scatter(history, history_vals, color='r', zorder=5)
plt.plot(history, history_vals, color='r', linestyle='--', label='Gradient Descent Steps')

# Mark the starting and minimum points
plt.scatter(initial_x, func(initial_x), color='green', label='Start (x=5)')
plt.scatter(-2, func(-2), color='purple', label='Minimum (x=-2)')

# Annotate all gradient descent steps
for i, x_pos in enumerate(history):
    plt.annotate(f"x={x_pos:.2f}\n f(x)={func(x_pos):.2f}", 
                 (x_pos, func(x_pos)), 
                 textcoords="offset points", 
                 xytext=(0, 5), ha='center', fontsize=8, color='black')

# Labels and Title
plt.title('Gradient Descent on f(x) = x^2 + 4x + 4')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()

plt.grid(True)
plt.show()
```
