```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
# Input data
x = np.linspace(-1, 1, 100)
y = 2 * x + 1 + np.random.normal(0, 0.1, size=x.shape)  # Add some noise

# Plot data
plt.scatter(x, y)
plt.title("Input Data")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

```python
model_deep = keras.Sequential([
    layers.Dense(64, input_shape=(1,), activation='relu'),  # First hidden layer with ReLU
    layers.Dropout(0.2),  # Dropout regularization (20% of the neurons are randomly dropped)
    layers.Dense(64, activation='relu'),  # Second hidden layer
    layers.Dropout(0.2),
    layers.Dense(32, activation='relu'),  # Third hidden layer
    layers.Dense(1)  # Output layer
])
```

```python
model_deep.compile(optimizer='adam', loss='mean_squared_error')
```

```python
history_deep = model_deep.fit(x, y, epochs=100, verbose=0)
```

```python
plt.plot(history_deep.history['loss'])
plt.title("Training Loss for Deep Neural Network")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.grid(True)
plt.show()
```
