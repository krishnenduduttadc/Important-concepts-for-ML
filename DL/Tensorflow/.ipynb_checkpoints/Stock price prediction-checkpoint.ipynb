```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

# 1. Create synthetic stock closing price data (sine wave + noise to simulate real stock)
np.random.seed(42)
days = np.arange(200)
prices = np.sin(0.1 * days) * 20 + 100 + np.random.normal(0, 1, size=200)  # Fake prices around $100

# 2. Prepare input sequences (last 5 days → next day)
def create_dataset(series, window_size=5):
    X, Y = [], []
    for i in range(len(series) - window_size):
        X.append(series[i:i + window_size])
        Y.append(series[i + window_size])
    return np.array(X, dtype=np.float32), np.array(Y, dtype=np.float32).reshape(-1, 1)

X, y = create_dataset(prices)

# 3. Build a simple neural network model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=(5,)),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(1)
])

# 4. Compile and train
model.compile(optimizer='adam', loss='mse')
model.fit(X, y, epochs=100, verbose=0)



# 5. Optional: Plot predictions vs actual
y_pred = model.predict(X)

plt.figure(figsize=(10, 4))
plt.plot(y, label='Actual')
plt.plot(y_pred, label='Predicted')
plt.title('Stock Price Prediction (Synthetic Data)')
plt.legend()
plt.grid(True)
plt.show()


# 6. Predict the next value based on the last 5 days
last_5_days = prices[-5:].reshape(1, -1).astype(np.float32)
predicted_price = model.predict(last_5_days)
print("Next day predicted price:", predicted_price[0][0])
```
