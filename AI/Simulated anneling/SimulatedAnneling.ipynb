```python
import numpy as np
import matplotlib.pyplot as plt
import time

# Objective function
def objective_function(x):
    return x**2 + 10 * np.sin(x)

# Simulated Annealing
def simulated_annealing(objective, bounds, max_iter, initial_temp, alpha):
    x = np.random.uniform(bounds[0], bounds[1])
    best_x = x
    best_eval = objective(x)
    current_temp = initial_temp

    # Store results for visualization
    history = [(x, best_eval)]

    for i in range(max_iter):
        # Candidate solution
        candidate_x = x + np.random.uniform(-1, 1)
        candidate_x = np.clip(candidate_x, bounds[0], bounds[1])
        candidate_eval = objective(candidate_x)

        # Difference in evaluation
        delta = candidate_eval - objective(x)

        # Accept candidate?
        if delta < 0 or np.random.rand() < np.exp(-delta / current_temp):
            x = candidate_x
            if candidate_eval < best_eval:
                best_x, best_eval = candidate_x, candidate_eval

        # Store for plotting
        history.append((x, objective(x)))

        # Cool down
        current_temp *= alpha

    return best_x, best_eval, history

# Plot function and steps
def plot_simulation(history, objective, bounds):
    x_vals = np.linspace(bounds[0], bounds[1], 1000)
    y_vals = objective(x_vals)

    plt.figure(figsize=(10, 6))
    for i, (x, y) in enumerate(history):
        plt.clf()
        plt.plot(x_vals, y_vals, label='Objective Function')
        plt.scatter([x], [y], color='red')
        plt.title(f"Step {i+1}: x = {x:.4f}, f(x) = {y:.4f}")
        plt.xlabel('x')
        plt.ylabel('f(x)')
        plt.pause(0.1)

    plt.show()

# Run simulation
bounds = [-10, 10]
max_iter = 100
initial_temp = 10
cooling_rate = 0.95

best_x, best_eval, history = simulated_annealing(objective_function, bounds, max_iter, initial_temp, cooling_rate)

print(f"Best solution: x = {best_x:.4f}, f(x) = {best_eval:.4f}")

# Visualization
plot_simulation(history, objective_function, bounds)
```
