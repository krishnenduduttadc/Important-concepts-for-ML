```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random cities
def generate_cities(num_cities, seed=42):
    np.random.seed(seed)
    return np.random.rand(num_cities, 2) * 100  # Coordinates (x, y)

# Distance matrix
def calculate_distances(cities):
    num = len(cities)
    dist = np.zeros((num, num))
    for i in range(num):
        for j in range(num):
            dist[i, j] = np.linalg.norm(cities[i] - cities[j])
    return dist

# Greedy TSP algorithm
def greedy_tsp(dist_matrix):
    n = len(dist_matrix)
    visited = [0]
    current = 0
    while len(visited) < n:
        next_city = np.argmin([dist_matrix[current, j] if j not in visited else np.inf for j in range(n)])
        visited.append(next_city)
        current = next_city
    visited.append(0)  # Return to start
    return visited

# Visualization
def visualize_tour(cities, tour, pause=0.8):
    plt.clf()
    x, y = cities[:, 0], cities[:, 1]
    plt.scatter(x, y, c='red', s=50)
    for i, (xi, yi) in enumerate(zip(x, y)):
        plt.text(xi + 1, yi + 1, str(i), fontsize=9)

    for i in range(len(tour) - 1):
        start = cities[tour[i]]
        end = cities[tour[i + 1]]
        plt.plot([start[0], end[0]], [start[1], end[1]], 'b-o')
        plt.title(f"Step {i+1}/{len(tour)-1}")
        plt.pause(pause)

    plt.title(f"Final Tour Length: {calculate_tour_length(cities, tour):.2f}")
    plt.show()

# Total tour length
def calculate_tour_length(cities, tour):
    return sum(np.linalg.norm(cities[tour[i]] - cities[tour[i + 1]]) for i in range(len(tour) - 1))

# Run everything
cities = generate_cities(num_cities=10)
dist_matrix = calculate_distances(cities)
tour = greedy_tsp(dist_matrix)
visualize_tour(cities, tour)
```
