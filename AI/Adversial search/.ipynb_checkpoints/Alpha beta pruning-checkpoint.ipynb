```python
import matplotlib.pyplot as plt
import networkx as nx
import time

# Define the tree manually
# Format: (node_name, value or None, children_names)
tree = {
    "A": (None, ["B", "C"]),
    "B": (None, ["D", "E"]),
    "C": (None, ["F", "G"]),
    "D": (3, []),
    "E": (5, []),
    "F": (6, []),
    "G": (9, [])
}

# Track graph and values
G = nx.DiGraph()
values = {}

# Build the graph for visualization
def build_graph():
    for node, (val, children) in tree.items():
        G.add_node(node, value=val)
        for child in children:
            G.add_edge(node, child)

# Minimax with Alpha-Beta
def minimax(node, depth, alpha, beta, maximizing):
    val, children = tree[node]
    if val is not None:
        print(f"Visited leaf {node} with value {val}")
        return val

    if maximizing:
        max_eval = -float("inf")
        for child in children:
            eval = minimax(child, depth + 1, alpha, beta, False)
            max_eval = max(max_eval, eval)
            alpha = max(alpha, eval)
            print(f"MAX node {node}: alpha={alpha}, beta={beta}")
            if beta <= alpha:
                print(f"Pruned at MAX node {node}")
                break
        values[node] = max_eval
        return max_eval
    else:
        min_eval = float("inf")
        for child in children:
            eval = minimax(child, depth + 1, alpha, beta, True)
            min_eval = min(min_eval, eval)
            beta = min(beta, eval)
            print(f"MIN node {node}: alpha={alpha}, beta={beta}")
            if beta <= alpha:
                print(f"Pruned at MIN node {node}")
                break
        values[node] = min_eval
        return min_eval

# Visualization step
def draw_tree(highlight_node=None):
    pos = nx.spring_layout(G)
    plt.clf()
    node_labels = {n: f"{n}\n{values[n] if n in values else ''}" for n in G.nodes}
    color_map = ['lightgreen' if n == highlight_node else 'lightblue' for n in G.nodes]

    nx.draw(G, pos, with_labels=True, labels=node_labels, node_color=color_map, node_size=1200, font_size=10)
    plt.pause(1)

# Run everything
build_graph()
plt.figure(figsize=(8, 5))
draw_tree()
result = minimax("A", 0, -float("inf"), float("inf"), True)
draw_tree()
print(f"\nFinal result at root A: {result}")
plt.show()
```
