```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# 1. Generate data: y = x^2 + 3
x = np.linspace(1, 10, 7000).reshape(-1, 1).astype(np.float32)

y = (x ** 2 + 3).astype(np.float32)


# 2. Build the model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(10, activation='relu', input_shape=(1,)),
    tf.keras.layers.Dense(1)
])

# 3. Compile the model with a smaller learning rate
optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)  # Lower learning rate
model.compile(optimizer=optimizer, loss='mse')

# 4. Train the model
history = model.fit(x, y, epochs=300, verbose=0)

# 5. Predict and plot
y_pred = model.predict(x)

plt.plot(x, y, label='True Function')
plt.plot(x, y_pred, label='Model Prediction')
plt.legend()
plt.title("Approximating y = x² + 3 with TensorFlow")
plt.show()

pred = model.predict(np.array([[5]]).astype(np.float32))
print("Prediction for x = 5:", pred)
```
