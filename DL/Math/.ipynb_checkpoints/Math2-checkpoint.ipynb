```python
import numpy as np
import matplotlib.pyplot as plt

# Sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Derivative of sigmoid
def sigmoid_derivative(x):
    s = sigmoid(x)
    return s * (1 - s)

# x values
x = np.linspace(-10, 10, 400)

# y values
y = sigmoid(x)
y_derivative = sigmoid_derivative(x)

# Plot
plt.figure(figsize=(8, 5))
plt.plot(x, y, label='Sigmoid Function')
plt.plot(x, y_derivative, label='Sigmoid Derivative')
plt.axhline(0, color='gray', linestyle='--')
plt.legend()
plt.title('Sigmoid Function and Its Derivative')
plt.grid(True)
plt.show()
```

```python
# Input:
# 100x100x1
# (grayscale image)
# Conv layer:
# 10
# filters of
# 3x3, stride1, padding1.
# Max Pooling:
# 2x2, stride2.
# What is the output shape after these two layer
```

```python
def conv_output_size(input_size, filter_size, stride, padding):
    return (input_size - filter_size + 2 * padding) // stride + 1

def pool_output_size(input_size, filter_size, stride):
    return (input_size - filter_size) // stride + 1

# Input dimensions
input_height = input_width = 100
input_depth = 1

# Convolution parameters
conv_filter_size = 3
conv_stride = 1
conv_padding = 1
num_filters = 10

# Convolution output
conv_height = conv_output_size(input_height, conv_filter_size, conv_stride, conv_padding)
conv_width = conv_output_size(input_width, conv_filter_size, conv_stride, conv_padding)
conv_depth = num_filters

# Max pooling parameters
pool_filter_size = 2
pool_stride = 2

# Pooling output
pool_height = pool_output_size(conv_height, pool_filter_size, pool_stride)
pool_width = pool_output_size(conv_width, pool_filter_size, pool_stride)
pool_depth = conv_depth

# Final output shape
output_shape = (pool_height, pool_width, pool_depth)
print(f"Output shape after Conv + Pooling: {output_shape}")
```

```python
# A CNN layer processes an input of size
# (64, 64, 16) with 32 filters of size (7, 7), stride 1 and no padding. How manymultiplications are required for this layer?
```

```python
input_height = input_width = 64
input_channels = 16
filter_height = filter_width = 7
num_filters = 32
stride = 1
padding = 0

# Output dimensions
output_height = (input_height - filter_height + 2 * padding) // stride + 1
output_width = (input_width - filter_width + 2 * padding) // stride + 1

# Multiplications per output element
mults_per_filter = filter_height * filter_width * input_channels

# Total multiplications
total_multiplications = output_height * output_width * num_filters * mults_per_filter

print(f"Total multiplications: {total_multiplications:,}")
```

```python
# Input:
# (32, 32, 3)

# Architecture:
# Two Conv2D layers:
# 1. 32 filters of (5, 5) stride 1 , padding 2

# 2. 64 filters of (3, 3) , stride 2, padding 0.
# What is the total number of trainable parameters
```

```python
def conv2d_params(kernel_size, in_channels, out_channels, bias=True):
    k_h, k_w = kernel_size
    weights = k_h * k_w * in_channels * out_channels
    biases = out_channels if bias else 0
    return weights + biases

# Layer 1
layer1_params = conv2d_params((5, 5), in_channels=3, out_channels=32)

# Layer 2
layer2_params = conv2d_params((3, 3), in_channels=32, out_channels=64)

# Total
total_params = layer1_params + layer2_params
print(f"Total trainable parameters: {total_params}")
```

```python
# Input:
# (224, 224, 3)

# Layer:
# Conv2D with 64 filters of (7, 7) , stride 1, padding '3
# , followed by MaxPooling (2, 2),stride 2.
# Total multiplications required for the above layers?
```

```python
# Given values
input_height = input_width = 224
input_channels = 3
num_filters = 64
kernel_height = kernel_width = 7
stride_conv = 1
padding_conv = 3

# Output size of Conv2D layer (same as input size for padding=3)
output_height_conv = (input_height - kernel_height + 2 * padding_conv) // stride_conv + 1
output_width_conv = output_height_conv  # Same as height
output_depth_conv = num_filters

# Multiplications for Conv2D layer
mults_conv2d = output_height_conv * output_width_conv * output_depth_conv * (kernel_height * kernel_width * input_channels)

# Given MaxPooling layer parameters
pool_size = 2
stride_pool = 2

# Output size of MaxPooling layer
output_height_pool = output_width_pool = output_height_conv // pool_size
output_depth_pool = output_depth_conv  # Depth remains the same

# Total multiplications
total_multiplications = mults_conv2d

print(f"Total multiplications for Conv2D and MaxPooling: {total_multiplications:,}")
```

```python
#In a CNN with 3 convolutional layers, each having 256 filters of size 3x3 and a stride of 1, followed by max-pooling layers with a pool size of 2x2 (stride: 2), if the input image size is 224x224x3, what will be the output sizeof the feature map after the third convolutional layer? Assume no padding is used.
```

```python
def calc_output_size(input_size, filter_size, stride, padding=0):
    return (input_size - filter_size + 2 * padding) // stride + 1

# Initial input size
input_size = 224
num_filters = 256
filter_size = 3
stride = 1

# After first conv + pool
output_size = calc_output_size(input_size, filter_size, stride)
output_size_pool = output_size // 2

# After second conv + pool
output_size = calc_output_size(output_size_pool, filter_size, stride)
output_size_pool = output_size // 2

# After third conv + pool
output_size = calc_output_size(output_size_pool, filter_size, stride)
output_size_pool = output_size // 2

print(f"Output size after third convolution and pooling: {output_size_pool}x{output_size_pool}x{num_filters}")
```

```python
# What is the receptive field of a pixel after three 3×3 conv layers (stride=1)?
```

```python
def calculate_receptive_field(layers, kernel_size, stride):
    receptive_field = 1  # Initially, the receptive field is the pixel itself
    for _ in range(layers):
        receptive_field += (kernel_size - 1) * stride
    return receptive_field

# Given values
kernel_size = 3
stride = 1
layers = 3

# Calculate receptive field
receptive_field = calculate_receptive_field(layers, kernel_size, stride)
print(f"Receptive field after {layers} convolutional layers: {receptive_field}x{receptive_field}")
```

```python
# A fully connected neural network (FCNN) has two fully connected layers. The fi rst layer has 50 output neurons connected to 128 input neurons, and the second layer has 30 outputs neurons connected to the 50neurons from the previous layer. Each neuron in both layers also has a bias. How many weights (including bias weights) are being updated in total across both layers during training?
```

```python
# Given values
input_neurons_layer1 = 128
output_neurons_layer1 = 50
input_neurons_layer2 = 50
output_neurons_layer2 = 30

# Calculate weights and biases for layer 1
weights_layer1 = input_neurons_layer1 * output_neurons_layer1
biases_layer1 = output_neurons_layer1

# Calculate weights and biases for layer 2
weights_layer2 = input_neurons_layer2 * output_neurons_layer2
biases_layer2 = output_neurons_layer2

# Total weights (including biases) across both layers
total_weights = (weights_layer1 + biases_layer1) + (weights_layer2 + biases_layer2)
print(f"Total weights (including bias weights): {total_weights}")
```

```python
# Consider a fully connected neural network (FCNN) with the following architecture:
# Hidden layer 1: 512 neurons, using ReLU activation
# Hidden layer 2: 256 neurons, using ReLU activation
# Output layer: 10 neurons (for digit classifi cation, 0-9), using softmax activation
# If the input to the network is a single 28x28 image, what is the total number of trainable parameters in this FCNN?
```

```python
# Given values
input_neurons = 784  # Flattened 28x28 image
hidden_layer1_neurons = 512
hidden_layer2_neurons = 256
output_neurons = 10

# Number of parameters for each layer
weights_layer1 = input_neurons * hidden_layer1_neurons
biases_layer1 = hidden_layer1_neurons

weights_layer2 = hidden_layer1_neurons * hidden_layer2_neurons
biases_layer2 = hidden_layer2_neurons

weights_output = hidden_layer2_neurons * output_neurons
biases_output = output_neurons

# Total trainable parameters
total_params = (weights_layer1 + biases_layer1) + (weights_layer2 + biases_layer2) + (weights_output + biases_output)
print(f"Total trainable parameters: {total_params}")
```
