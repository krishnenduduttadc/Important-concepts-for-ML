```python
import numpy as np
import matplotlib.pyplot as plt

# Vectors
a = np.array([1, 2])
b = np.array([3, 4])
dot_product = np.dot(a, b)

print("Dot Product:", dot_product)

# Matrix Multiplication
A = np.array([[1, 2], [3, 4]])
B = np.array([[2, 0], [1, 2]])
product = np.matmul(A, B)
print("Matrix Product:\n", product)
```

```python
#Activation Functions
```

```python
x = np.linspace(-10, 10, 100)

# Sigmoid
sigmoid = 1 / (1 + np.exp(-x))

# Tanh
tanh = np.tanh(x)

# ReLU
relu = np.maximum(0, x)

# Softmax
def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum()

softmax_vals = softmax(np.array([1, 2, 3]))

# Plot
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.plot(x, sigmoid, label='Sigmoid')
plt.plot(x, tanh, label='Tanh')
plt.plot(x, relu, label='ReLU')
plt.title("Activation Functions")
plt.legend()

plt.subplot(1, 3, 2)
plt.bar([1,2,3], softmax_vals)
plt.title("Softmax Output")

plt.tight_layout()
plt.show()
```

```python
#Loss function
```

```python
# Mean Squared Error
y_true = np.array([1, 0, 1])
y_pred = np.array([0.8, 0.1, 0.6])
mse = np.mean((y_true - y_pred) ** 2)
print("MSE:", mse)

# Cross-Entropy Loss (binary)
epsilon = 1e-12  # to avoid log(0)
cross_entropy = -np.mean(y_true * np.log(y_pred + epsilon) + (1 - y_true) * np.log(1 - y_pred + epsilon))
print("Binary Cross-Entropy:", cross_entropy)
```

```python
#Gradient descent
```

```python
# f(x) = x^2; f'(x) = 2x
x = 10  # Initial value
lr = 0.1  # Learning rate
history = []

for _ in range(20):
    grad = 2 * x
    x -= lr * grad
    history.append(x)

print("Minimum x:", x)

plt.plot(history)
plt.title("Gradient Descent on f(x) = x²")
plt.xlabel("Iterations")
plt.ylabel("x value")
plt.show()
```

```python
#Backpropagation Basics
```

```python
# Forward: z = w*x + b, a = sigmoid(z)
# Backward: dL/dw, dL/db

def sigmoid_deriv(z):
    s = 1 / (1 + np.exp(-z))
    return s * (1 - s)

x = 1.0
w = 0.5
b = 0.0

# Forward pass
z = w * x + b
a = 1 / (1 + np.exp(-z))  # sigmoid

# Suppose L = (a - y)^2
y = 1.0
dL_da = 2 * (a - y)
da_dz = sigmoid_deriv(z)
dz_dw = x

# Backward
dL_dw = dL_da * da_dz * dz_dw
print("Gradient of Loss w.r.t weight:", dL_dw)
```

```python
#Layer Output Size (CNN)

# W: Input size
# K: Kernel size
# P: Padding
# S: Stride
```

```python
def conv2d_output_size(w, k, p, s):
    return int((w - k + 2 * p) / s) + 1

# Example
input_size = 32
kernel_size = 3
padding = 1
stride = 1

output = conv2d_output_size(input_size, kernel_size, padding, stride)
print(f"Conv2D Output Size: {output}x{output}")
```

```python
def dense_params(input_features, output_features):
    return input_features * output_features + output_features

params = dense_params(784, 128)
print(f"Dense Layer Parameters: {params}")
```

```python
#ANN Forward Pass
```

```python
import numpy as np

# Example: 3-layer ANN
X = np.array([[0.5, 0.3]])
W1 = np.random.rand(2, 4)  # input to hidden
b1 = np.zeros((1, 4))
W2 = np.random.rand(4, 1)  # hidden to output
b2 = np.zeros((1, 1))

# Forward pass
hidden = np.maximum(0, np.dot(X, W1) + b1)  # ReLU
output = 1 / (1 + np.exp(-(np.dot(hidden, W2) + b2)))  # Sigmoid
print("ANN Output:", output)
```

```python
#Stride and Kernel Influence (Visualization)
```

```python
import matplotlib.pyplot as plt
import numpy as np

image = np.random.rand(10, 10)

def apply_kernel(img, kernel_size, stride):
    h, w = img.shape
    out_h = int((h - kernel_size) / stride) + 1
    out_w = int((w - kernel_size) / stride) + 1
    output = np.zeros((out_h, out_w))
    for i in range(0, h - kernel_size + 1, stride):
        for j in range(0, w - kernel_size + 1, stride):
            region = img[i:i+kernel_size, j:j+kernel_size]
            output[i//stride, j//stride] = np.mean(region)
    return output

out_img = apply_kernel(image, 3, 2)
plt.imshow(out_img, cmap='gray')
plt.title("Output with Kernel=3 and Stride=2")
plt.colorbar()
plt.show()
```

```python
# Deep Learning Math Formulas - Complete Python Examples

import numpy as np
import matplotlib.pyplot as plt

# -------------------------------
# 1. Weight Initialization
# -------------------------------
def xavier_init(n_in, n_out):
    return np.random.randn(n_in, n_out) * np.sqrt(1 / (n_in + n_out))

def he_init(n_in):
    return np.random.randn(n_in, 1) * np.sqrt(2 / n_in)

def example_weight_initialization():
    W_xavier = xavier_init(4, 3)
    W_he = he_init(4)
    print("Xavier Initialized Weights:\n", W_xavier)
    print("He Initialized Weights:\n", W_he)

# -------------------------------
# 2. Batch Normalization
# -------------------------------
def batch_norm(x, epsilon=1e-5):
    mu = np.mean(x, axis=0)
    var = np.var(x, axis=0)
    return (x - mu) / np.sqrt(var + epsilon)

def example_batch_norm():
    x_batch = np.random.rand(5, 3)
    x_norm = batch_norm(x_batch)
    print("Original Batch:\n", x_batch)
    print("Batch-Normalized:\n", x_norm)

# -------------------------------
# 3. Dropout
# -------------------------------
def dropout(x, drop_prob):
    mask = (np.random.rand(*x.shape) > drop_prob).astype(float)
    return x * mask / (1.0 - drop_prob)

def example_dropout():
    x_input = np.ones((4, 4))
    x_dropped = dropout(x_input, 0.5)
    print("Original Input:\n", x_input)
    print("After Dropout:\n", x_dropped)

# -------------------------------
# 4. Softmax Derivative
# -------------------------------
def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum()

def softmax_derivative(s):
    return np.diagflat(s) - np.dot(s, s.T)

def example_softmax_derivative():
    x = np.array([2.0, 1.0, 0.1])
    s = softmax(x)
    jacobian = softmax_derivative(s)
    print("Softmax Output:\n", s)
    print("Softmax Jacobian Matrix:\n", jacobian)

# -------------------------------
# 5. LSTM Cell
# -------------------------------
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def lstm_cell(x, h_prev, c_prev, Wf, Wi, Wc, Wo, bf, bi, bc, bo):
    f = sigmoid(np.dot(Wf, x) + bf)
    i = sigmoid(np.dot(Wi, x) + bi)
    c_ = np.tanh(np.dot(Wc, x) + bc)
    c = f * c_prev + i * c_
    o = sigmoid(np.dot(Wo, x) + bo)
    h = o * np.tanh(c)
    return h, c

def example_lstm_cell():
    x = np.random.randn(3, 1)
    h_prev = np.zeros((3, 1))
    c_prev = np.zeros((3, 1))
    Wf = np.random.randn(3, 3)
    Wi = np.random.randn(3, 3)
    Wc = np.random.randn(3, 3)
    Wo = np.random.randn(3, 3)
    bf = np.zeros((3, 1))
    bi = np.zeros((3, 1))
    bc = np.zeros((3, 1))
    bo = np.zeros((3, 1))
    h, c = lstm_cell(x, h_prev, c_prev, Wf, Wi, Wc, Wo, bf, bi, bc, bo)
    print("LSTM Hidden State:\n", h)
    print("LSTM Cell State:\n", c)

# -------------------------------
# 6. Adam Optimizer
# -------------------------------
def adam_update(param, grad, m, v, t, lr=0.001, beta1=0.9, beta2=0.999, epsilon=1e-8):
    m = beta1 * m + (1 - beta1) * grad
    v = beta2 * v + (1 - beta2) * (grad ** 2)
    m_hat = m / (1 - beta1 ** t)
    v_hat = v / (1 - beta2 ** t)
    param -= lr * m_hat / (np.sqrt(v_hat) + epsilon)
    return param, m, v

def example_adam_optimizer():
    param = np.array([1.0])
    grad = np.array([0.1])
    m = v = np.zeros(1)
    print("Initial Param:", param)
    for t in range(1, 6):
        param, m, v = adam_update(param, grad, m, v, t)
        print(f"After step {t}, Param: {param}")

# -------------------------------
# 7. Learning Rate Schedulers
# -------------------------------
def step_decay(epoch, init_lr=0.1, drop=0.5, epochs_drop=10):
    return init_lr * (drop ** (epoch // epochs_drop))

def exp_decay(epoch, init_lr=0.1, k=0.1):
    return init_lr * np.exp(-k * epoch)

def example_lr_schedulers():
    epochs = np.arange(0, 50)
    lr_step = [step_decay(e) for e in epochs]
    lr_exp = [exp_decay(e) for e in epochs]
    plt.plot(epochs, lr_step, label="Step Decay")
    plt.plot(epochs, lr_exp, label="Exp Decay")
    plt.xlabel("Epoch")
    plt.ylabel("Learning Rate")
    plt.title("Learning Rate Schedulers")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# -------------------------------
# Run all examples
# -------------------------------
example_weight_initialization()
example_batch_norm()
example_dropout()
example_softmax_derivative()
example_lstm_cell()
example_adam_optimizer()
example_lr_schedulers()
```
