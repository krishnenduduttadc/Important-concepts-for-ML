```python
import numpy as np
import matplotlib.pyplot as plt

# ReLU activation
def relu(x):
    return np.maximum(0, x)

# Settings
np.random.seed(42)
input_data = np.random.randn(1000, 100)  # 1000 samples, 100 features
num_layers = 20
layer_size = 100

# Activation variance lists
activations_std = [input_data]
activations_xavier = [input_data]
activations_he = [input_data]

# Standard Normal
x_std = input_data.copy()
for _ in range(num_layers):
    w = np.random.randn(layer_size, layer_size)
    x_std = relu(x_std @ w)
    activations_std.append(x_std)

# Xavier (Glorot)
x_xavier = input_data.copy()
for _ in range(num_layers):
    scale = np.sqrt(2.0 / (layer_size + layer_size))
    w = np.random.randn(layer_size, layer_size) * scale
    x_xavier = relu(x_xavier @ w)
    activations_xavier.append(x_xavier)

# He Initialization
x_he = input_data.copy()
for _ in range(num_layers):
    scale = np.sqrt(2.0 / layer_size)
    w = np.random.randn(layer_size, layer_size) * scale
    x_he = relu(x_he @ w)
    activations_he.append(x_he)

# Compute variances
var_std = [np.var(a) for a in activations_std]
var_xavier = [np.var(a) for a in activations_xavier]
var_he = [np.var(a) for a in activations_he]

# Plotting
plt.figure(figsize=(10, 5))
plt.plot(range(num_layers + 1), var_std, label="Standard Normal", marker='o', color='crimson')
plt.plot(range(num_layers + 1), var_xavier, label="Xavier Init", marker='x', color='blue')
plt.plot(range(num_layers + 1), var_he, label="He Init", marker='s', color='green')
plt.xlabel("Layer")
plt.ylabel("Activation Variance")
plt.title("Activation Variance Across Layers")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
