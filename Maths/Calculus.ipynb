```python
#diffarentiation f(x)=x^2 + 3*x + 5
```

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
# Define function
x = sp.Symbol('x')
f_expr = x**2 + 3*x + 5
f = sp.lambdify(x, f_expr, 'numpy') # Convert symbolic to numerical function
# Compute derivative
df_expr = sp.diff(f_expr, x)
df = sp.lambdify(x, df_expr, 'numpy')
# Generate x values
x_vals = np.linspace(-5, 5, 100)
y_vals = f(x_vals)
# Tangent at x=1
x_tangent = 1
y_tangent = f(x_tangent)
slope = df(x_tangent)
tangent_line = slope * (x_vals - x_tangent) + y_tangent
# Plot
plt.figure(figsize=(8, 5))
plt.plot(x_vals, y_vals, label=r'$f(x) = x^2 + 3x + 5$')
plt.plot(x_vals, tangent_line, '--', label=f'Tangent at x={x_tangent}')
plt.scatter([x_tangent], [y_tangent], color='red', zorder=3)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.title("Differentiation: Function and Tangent Line")
plt.grid()
plt.show()
```

```python
# Compute indefinite integral f(x)=x^2 + 3x + 5
```

```python
g_expr = x**2 + 3*x + 5
int_g_expr = sp.integrate(g_expr, x)
int_g = sp.lambdify(x, int_g_expr, 'numpy')
# Generate x values
x_vals = np.linspace(-5, 5, 100)
y_vals = f(x_vals)
y_int_vals = int_g(x_vals)
# Plot function and its integral
plt.figure(figsize=(8, 5))
plt.plot(x_vals, y_vals, label=r'$g(x) = x^2 + 3x + 5$')
plt.plot(x_vals, y_int_vals, '--', label=r'$\int g(x)dx$')
plt.xlabel('x')
plt.ylabel('Function Value')
plt.legend()
plt.title("Integration: Function and Indefinite Integral")
plt.grid()
plt.show()
```

```python
#Definite integral, area under curve for f(x)=x^2 + 3x + 5
```

```python
from scipy.integrate import quad
# Define function for integration
h = lambda x: x**2 + 3*x + 5
a, b = 0, 2 # Limits of integration
x_fill = np.linspace(a, b, 100)
y_fill = h(x_fill)
# Compute numerical integral
area, _ = quad(h, a, b)
# Plot function and shaded region
plt.figure(figsize=(8, 5))
plt.plot(x_vals, f(x_vals), label=r'$h(x) = x^2 + 3x + 5$')
plt.fill_between(x_fill, y_fill, color='cyan', alpha=0.5, label=f"Area = {area:.2f}")
plt.xlabel('x')
plt.ylabel('h(x)')
plt.legend()
plt.title("Definite Integral: Area Under Curve")
plt.grid()
plt.show()
```

```python
#Area under Straight line f(x)=2x+3
```

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
from scipy.integrate import quad
# Define symbols
x = sp.Symbol('x')
# Straight Line Function
f_expr = 2*x + 3
int_f_expr = sp.integrate(f_expr, x) # Indefinite Integral
# Convert to function
f = sp.lambdify(x, f_expr, 'numpy')
# Definite Integral from x=0 to x=2
area, _ = quad(f, 0, 2)
# Generate x values
x_vals = np.linspace(-2, 3, 100)
y_vals = f(x_vals)
x_fill = np.linspace(0, 2, 100)
y_fill = f(x_fill)
# Plot
plt.figure(figsize=(8, 5))
plt.plot(x_vals, y_vals, label=r'$f(x) = 2x + 3$')
plt.fill_between(x_fill, y_fill, color='cyan', alpha=0.5, label=f"Area = {area:.2f}")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.title("Integration of a Straight Line")
plt.grid()
plt.show()
```

```python
#Area under parabola f(x)=x^24x+3
```

```python
# Parabola Function
g_expr = x**2 - 4*x + 3
int_g_expr = sp.integrate(g_expr, x)
# Convert to function
g = sp.lambdify(x, g_expr, 'numpy')
# Definite Integral from x=1 to x=3
area, _ = quad(g, 1, 3)
# Generate values
x_vals = np.linspace(-1, 5, 100)
y_vals = g(x_vals)
x_fill = np.linspace(1, 3, 100)
y_fill = g(x_fill)
# Plot
plt.figure(figsize=(8, 5))
plt.plot(x_vals, y_vals, label=r'$g(x) = x^2 - 4x + 3$')
plt.fill_between(x_fill, y_fill, color='orange', alpha=0.5, label=f"Area ={area:.2f}")
plt.xlabel('x')
plt.ylabel('g(x)')
plt.legend()
plt.title("Integration of a Parabola")
plt.grid()
plt.show()
```

```python
#Area under circle x^2+y^2 =4
```

```python
# Circle Function (Upper semicircle of x^2 + y^2 = 4)
h = lambda x: np.sqrt(4 - x**2)
# Definite Integral from x=-2 to x=2
area, _ = quad(h, -2, 2)
# Generate values
x_vals = np.linspace(-2, 2, 100)
y_vals = h(x_vals)
# Plot
plt.figure(figsize=(8, 5))
plt.plot(x_vals, y_vals, label=r'$y = \sqrt{4 - x^2}$')
plt.fill_between(x_vals, y_vals, color='purple', alpha=0.5, label=f"Area ={area:.2f}")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title("Integration of a Semicircle")
plt.grid()
plt.show()
```

```python
#Area under Ellipse x^2/4+y^2/2=1
```

```python
# Ellipse Function (Upper half of x^2/4 + y^2/1 = 1)
e = lambda x: np.sqrt(1 - (x**2 / 4))
# Definite Integral from x=-2 to x=2
area, _ = quad(e, -2, 2)
# Generate values
x_vals = np.linspace(-2, 2, 100)
y_vals = e(x_vals)
# Plot
plt.figure(figsize=(8, 5))
plt.plot(x_vals, y_vals, label=r'$y = \sqrt{1 - x^2/4}$')
plt.fill_between(x_vals, y_vals, color='green', alpha=0.5, label=f"Area = {area:.2f}")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title("Integration of an Ellipse (Upper Half)")
plt.grid()
plt.show()
```
