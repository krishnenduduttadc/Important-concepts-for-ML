```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# Step 1: Create simple data (y = 2x + 1 with noise)
x = np.linspace(-1, 1, 100)
y = 2 * x + 1 + np.random.normal(0, 0.1, size=x.shape)  # Add some noise

# Step 2: Build the Autoencoder

# Encoder part
encoder = keras.Sequential([
    layers.Dense(32, input_shape=(1,), activation='relu'),  # First encoding layer
    layers.Dense(16, activation='relu')  # Second encoding layer
])

# Decoder part
decoder = keras.Sequential([
    layers.Dense(32, input_shape=(16,), activation='relu'),  # First decoding layer
    layers.Dense(1)  # Output layer to reconstruct the input data
])

# Full autoencoder (combining encoder and decoder)
autoencoder = keras.Sequential([encoder, decoder])

# Step 3: Compile the Autoencoder Model
autoencoder.compile(optimizer='adam', loss='mean_squared_error')

# Step 4: Train the Autoencoder (using x as both input and target)
history_autoencoder = autoencoder.fit(x, x, epochs=100, verbose=0)

# Step 5: Plot the Training Loss
plt.plot(history_autoencoder.history['loss'])
plt.title("Training Loss for Autoencoder Model")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.grid(True)
plt.show()

# Step 6: Use the Autoencoder to Reconstruct the Input
x_reconstructed = autoencoder.predict(x)

# Step 7: Plot the Original Data and the Autoencoder Reconstruction
plt.plot(x, y, label="Original data")
plt.plot(x, x_reconstructed, color='red', label="Autoencoder reconstruction")
plt.title("Autoencoder Reconstruction")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```
