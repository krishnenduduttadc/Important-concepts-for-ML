```python
#AP
```

```python
import numpy as np
import matplotlib.pyplot as plt
# Define AP sequence
a, d, n = 2, 3, 10 # First term = 2, common difference = 3, 10 terms
AP = [a + (i * d) for i in range(n)]
# Plot
plt.figure(figsize=(8, 5))
plt.plot(range(1, n+1), AP, marker='o', linestyle='--', color='b', label="APSequence")
plt.xlabel("n (Term Number)")
plt.ylabel("Value")
plt.title("Arithmetic Progression (AP)")
plt.legend()
plt.grid()
plt.show()
```

```python
#GP
```

```python
# Define GP sequence
a, r, n = 2, 2, 30 # First term = 2, common ratio = 2, 10 terms
GP = [a * (r**i) for i in range(n)]
# Plot
plt.figure(figsize=(8, 5))
plt.plot(range(1, n+1), GP, marker='o', linestyle='--', color='g', label="GPSequence")
plt.xlabel("n (Term Number)")
plt.ylabel("Value")
plt.title("Geometric Progression (GP)")
plt.legend()
plt.grid()
plt.show()
```

```python
#AP sum, GP sum
```

```python
# AP Sum
AP_sum = [sum([a + (i * d) for i in range(k+1)]) for k in range(n)]
# GP Sum
GP_sum = [sum([a * (r**i) for i in range(k+1)]) for k in range(n)]
# Plot AP and GP Sum
plt.figure(figsize=(8, 5))
plt.plot(range(1, n+1), AP_sum, marker='o', linestyle='--', color='b',label="AP Sum")
plt.plot(range(1, n+1), GP_sum, marker='o', linestyle='--', color='g',label="GP Sum")
plt.xlabel("n (Number of Terms)")
plt.ylabel("Summation")
plt.title("Summation of AP and GP")
plt.legend()
plt.grid()
plt.show()
```

```python
#Permutation
```

```python
from math import factorial
import numpy as np
import matplotlib.pyplot as plt
# Define n values
n_values = np.arange(1, 11) # n from 1 to 10
r = 3 # Fixed r value
# Compute Permutations (Only if n >= r)
permutations = [factorial(n) // factorial(n - r) if n >= r else 0 for n in n_values]
# Compute Combinations (Only if n >= r)
combinations = [factorial(n) // (factorial(r) * factorial(n - r)) if n >= r else 0 for n in n_values]
# Plot Permutations
plt.figure(figsize=(8, 5))
plt.plot(n_values, permutations, marker='o', linestyle='--', color='r',label="P(n, 3)")
plt.xlabel("n (Total Items)")
plt.ylabel("Permutations")
plt.title("Permutation P(n, r) with r=3")
plt.legend()
plt.grid()
plt.show()
```

```python
#Combination
```

```python
combinations = [factorial(n) // (factorial(r) * factorial(n - r)) if n >= r else 0 for n in n_values]
# Plot Combinations
plt.figure(figsize=(8, 5))
plt.plot(n_values, combinations, marker='o', linestyle='--', color='m',label="C(n, 3)")
plt.xlabel("n (Total Items)")
plt.ylabel("Combinations")
plt.title("Combination C(n, r) with r=3")
plt.legend()
plt.grid()
plt.show()
```

```python
#relu  f(x)=max(0,x)
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-10, 10, 100)
relu = np.maximum(0, x)
plt.figure(figsize=(6,4))
plt.plot(x, relu, label="ReLU", color="b")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("ReLU Activation Function")
plt.axhline(0, color='black', lw=1)
plt.axvline(0, color='black', lw=1)
plt.legend()
plt.grid()
plt.show()
```

```python
#softmax
```

```python
def softmax(x):
    exp_x = np.exp(x - np.max(x)) # Subtract max to prevent overflow
    return exp_x / np.sum(exp_x)
    
x_values = np.linspace(-2, 2, 10)
softmax_values = softmax(x_values)

plt.figure(figsize=(6,4))
plt.bar(x_values, softmax_values, label="Softmax", color="orange")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Softmax Function")
plt.legend()
plt.grid()
plt.show()
```

```python
#Fibonacci
```

```python
def fibonacci(n):
    fib = [0, 1]
    for i in range(2, n):
        fib.append(fib[-1] + fib[-2])
    return fib
fib = fibonacci(20)
plt.plot(fib, marker="o", label="Fibonacci Sequence")
plt.title("Fibonacci Sequence")
plt.legend()
plt.grid()
plt.show()
```

```python
#Fourier series
```

```python
def fourier_series(x, n_terms):
    return sum(np.sin((2*k - 1) * x) / (2*k - 1) for k in range(1, n_terms + 1))
x = np.linspace(-np.pi, np.pi, 400)
y = fourier_series(x, 5)
plt.plot(x, y, label="Fourier Series Approximation")
plt.title("Fourier Series Approximation")
plt.legend()
plt.grid()
plt.show()
```
